# IMPORTANDO DADOS

```python
import pandas as pd
path= "/content/drive/MyDrive/Colab Notebooks/recomendacao"
```

```python
filmes = pd.read_csv(path+'/movies.csv')
filmes.columns = ["filmeId", "titulo", "genero"]
filmes = filmes.set_index('filmeId')
filmes.head()
```

```python
# links = pd.read_csv(path+'/links.csv')
# links.head()
```

```python
notas = pd.read_csv(path+'/ratings.csv')
notas.columns = ["usuarioId", "filmeId", "nota", "momento"]
notas.head()
```

```python
notas.describe()
```

# RECOMENDAR FILMES MAIS POPULARES

```python
total_de_votos = notas["filmeId"].value_counts()
filmes['total_de_votos'] = total_de_votos
filmes.head()
```

```python
filmes.sort_values("total_de_votos", ascending=False).head(5)
```

```python
notas_medias = notas.groupby("filmeId").mean()["nota"]
filmes["nota_media"] = notas_medias
filmes.sort_values("total_de_votos", ascending=False).head(5)
```
